```python
import numpy as np
import os 
import pandas as pd
import sys
import matplotlib
import math
```

```python
def pred_func(K,theta,rho):
    if len(theta)==2*K+1:
        b=theta[-1]
        w=theta[:K]
        phi=theta[K:2*K]
        fx=[]
        for x in range(101):
            val=0
            for k in range(K):
                val+=w[k]*math.cos((2*math.pi*x/rho[k])-phi[k])
            fx.append(b+val)
        return fx
```

##Q5a

```python
import matplotlib.pyplot as plt
x=[i for i in range(101)]
K=2
rho=[50,25]
theta=[0.5,0.75,-0.25,0.75,1]
fx=pred_func(K,theta,rho)
plt.plot(x,fx)
plt.title("f_theta curve plot")
plt.xlabel("X")
plt.ylabel("F_X")
plt.show()
```

```python
pts=[(20,0),(40,2.5)]
loss=0
for i in range(len(pts)):
    loss+=(pts[i][1]-fx[pts[i][0]])**2
loss=loss/len(pts)
print(loss)
```
